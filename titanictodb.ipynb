```python
class titanic:
    def __init__(self):
        self.conn = pymysql.connect(host='localhost', user='root', password='1111', db='titanic', charset='utf8')
        self.cur = self.conn.cursor()
        sql = 'create table if not exists titanic (sex varchar(10), age float, embarked varchar(10))'
        self.cur.execute(sql)
        self.conn.commit()
    
    def __del__(self):
        self.conn.close()
        
    def call_df(self):
        df = pd.read_csv('train.csv')
        df = df[['Sex','Age','Embarked']].dropna()
        df = df.reset_index(drop=True)
        return df
    
    def update(self):
        data = self.call_df()
        for idx in range(len(data.Sex)):
            sex = data['Sex'][idx]
            age = data['Age'][idx]
            embarked = data['Embarked'][idx]
            print(sex, age, embarked)
            sql = f"replace into titanic (sex, age, embarked) values('{sex}','{age}','{embarked}')"
            self.cur.execute(sql)
        self.conn.commit()



bbb = titanic()
a = bbb.call_df()
print(a.Embarked)    
bbb.update()
```
